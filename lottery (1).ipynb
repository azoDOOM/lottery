```python
from cProfile import label
import tkinter as tk
from turtle import title
from typing import Container
from tkinter import font as tkfont
from tkinter.filedialog import askopenfile 
import csv
import random
from tkinter import*

top=Tk()
top.title("Random picker")
top.minsize(600,600)
ent=Label(top,text="Enter the number of winners")
ent.pack()
num =Entry()
num.pack()
global file
def open_file():
        
        file = (askopenfile(mode='r', filetypes=[('I', '*csv')]))
        if file is not None:
            pass
        return file

info=Label(top,text="Select file")
info.pack()



def change():
        file = (askopenfile(mode='r', filetypes=[('I', '*csv')]))
        if file is not None:
            pass
        
        reader = csv.reader(file)
        for index, row in enumerate(reader):
            if index == 0:
                chosen_row = row
            else:
                r = random.randint(0, index)
                if r == 0:
                    chosen_row = row
            mystring=' '
            for x in chosen_row:
                mystring += ' '+x
                    
        show=Label(top,text=mystring)
        show.pack()

sub = Button (top, text="Submit",
                         command=change)
sub.pack()





top.mainloop()
```
